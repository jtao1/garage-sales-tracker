```python
import folium
from scraper import *

listings = read_file()

map = folium.Map(location=[41.74328471, -88.10861947], zoom_start=11)
folium.Marker([41.74328471, -88.10861947], tooltip='home', icon=folium.Icon(color='red')).add_to(map)
for listing in listings:
	tooltip = f'{listing.date_start}, {listing.date_end}'
	popup_data = f'{listing.street}, {listing.city} {listing.times}' 
	popup = folium.Popup(popup_data, max_width=180)
	coords = [listing.coords[0], listing.coords[1]]
	folium.Marker(coords, popup=popup, tooltip=tooltip).add_to(map)

map
```
